```python
import os
import pandas as pd
input_csv_path = os.path.join("E4-CA.csv")
input_data_set = pd.read_csv(input_csv_path)
input_data_set.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(input_data_set, test_size = 0.2, random_state = 42)

print("Data set length: ", len(input_data_set))
print("Train set length: ", len(train_set), " Percentage of orig. data set length: ", (float(len(train_set)) / len(input_data_set)))
print("Test set length: ", len(test_set), " Percentage of orig. data set length: ", (float(len(test_set)) / len(input_data_set)))
```

```python
def get_value_label_matrices(data_set):
    labels = data_set["Label"].copy().values
    values = data_set[[
        "Amplitude-X",
        "Median-X",
        "Mean-X",
        "Max-X",
        "Min-X",
        "Peak-to-Peak-X",
        "Variance-X",
        "Std Dev-X",
        "RMS-X",
        "Skewness-X",
        "Amplitude-Y",
        "Median-Y",
        "Mean-Y",
        "Max-Y",
        "Min-Y",
        "Peak-to-Peak-Y",
        "Variance-Y",
        "Std Dev-Y",
        "RMS-Y",
        "Skewness-Y",
        "Amplitude-Z",
        "Median-Z",
        "Mean-Z",
        "Max-Z",
        "Min-Z",
        "Peak-to-Peak-Z",
        "Variance-Z",
        "Std Dev-Z",
        "RMS-Z",
        "Skewness-Z"
    ]].copy().values
    
    # Sanity check ... the number of value / label records should match
    print("Parsed ", len(values), " values and ", len(labels), " labels")
    return values, labels
```

```python
training_values, training_labels = get_value_label_matrices(train_set)
```

```python
test_values, test_labels = get_value_label_matrices(test_set)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfClf = RandomForestClassifier(n_estimators=100)

rfClf = rfClf.fit(training_values, training_labels)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix

def generate_predictions_and_cross_validate(clf, values, labels):
    validator_num = 10 # number of validators to use ... feel free to play with this
    
    # generate the predicted labels with the provided classifer ...
    predicted_labels = cross_val_predict(clf, values, labels, cv = validator_num)
    
    # what predictions did the classifier generate?
    actual_predictions = confusion_matrix(labels, predicted_labels)
    print("Actual Predictions: ")
    print(actual_predictions)
    
    # what predictions *SHOULD* the classifier have generated?
    perfect_predictions = confusion_matrix(labels, labels)
    print("Perfect Predictions: ")
    print(perfect_predictions)
    
    # Let's see how we did ...
    score = cross_val_score(clf, values, labels, cv = validator_num, scoring = "accuracy")
    print("Classifier Accuracy: ", score)
    return predicted_labels
```

```python
predicted_labels = generate_predictions_and_cross_validate(rfClf, training_values, training_labels)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(training_labels, predicted_labels)
print("Precision: ", precision)

recall = recall_score(training_labels, predicted_labels)
print("Recall: ", recall)

fscore = f1_score(training_labels, predicted_labels)
print("F Score: ", fscore)
```

```python
# Once you have the precision, recall, and f-score where you want them,
# you can evaluate the test set as well!

predicted_test_labels = generate_predictions_and_cross_validate(rfClf, test_values, test_labels)

test_precision = precision_score(test_labels, predicted_test_labels)
print("Test Precision: ", test_precision)

test_recall = recall_score(test_labels, predicted_test_labels)
print("Test Recall: ", test_recall)

test_fscore = f1_score(test_labels, predicted_test_labels)
print("Test F-Score: ", test_fscore)
```
